# Problem
The velocity of an object is given by:
$$v_x(t)=e^{-t^2/100}(t+10\sin \pi t)$$
Both $v_x$ and $x$ are zero when $t=0$.

1. Numerically find the first time when $v_x$ equals zero after the start and find the distance traveled from the origin.
2. Find the final position of the object as $t \to \infty$.

```python
# Imports
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# Function v_x(t)

def calc_vx(t):
    return np.exp(-t**2/100)*(t+10*np.sin(np.pi*t))
```

## Part 1

```python
dt = 0.0001
vx, t = 0, 0

while vx >= 0:
    vx = calc_vx(t)
    t += dt

print("It takes %s seconds" % t)
```

## Part 2
Before we start part 2, let's analyze the overall behaviour. The overall dominant effect will be the decaying exponential function, with a time constant of:
$$\tau = \sqrt{100}=10 \text{ s}$$
so the time it takes for the mass to stop moving is on the order of magnitude of $\tau$. For our case, we'll let it be $10\tau=100 \text{ s}$. The graph plotted below confirms it.

```python
x = np.linspace(0,30,1000) # Domain
y = calc_vx(x) # Function

# Create Plot
fig = plt.figure()
plt.plot(x,y, 'r')
plt.xlabel('$t$')
plt.ylabel('$v_x$')
plt.show()
```

```python
dt = 0.001
x, vx, t = 0, 0, 0

for t in np.arange(0,100,dt):
    vx = calc_vx(t)
    x += vx*dt
    t += dt

print("The final position as t goes to infinity is %s meters." % x)
```
