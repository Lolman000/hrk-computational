```python
import numpy as np
import matplotlib.pyplot as plt
```

After drawing a diagram and labeling the angle of incidence as $\theta$ from simple geometry and trigonometry it is obvious that $$\sin\theta=\frac{d}{R}.$$ Let the equation of the reflected ray be $y=mx+b$.The $m$ is $\tan 2\theta$ since the angle of incidence and angle of reflection are the same. To find $b$ we can use the incidence point. The coordinates of this point is $(R\cos\theta,d)$. Plugging it to equation yields with $b=d-R\cos\theta\tan 2\theta$. To find where it will cross the axis we need to set value of $y$ to $0$. After doing that for any value of $d$ the point of crossing axis is $$x=\frac{R\cos\left(\arcsin\left(\frac{d}{R}\right)\right)\tan\left(2\arcsin\left(\frac{d}{R}\right)\right)-d}{\tan\left(2\arcsin\left(\frac{d}{R}\right)\right)}$$
Implementing it to python is pretty simple

```python
def cross_value(d):
    numerator = np.cos(np.arcsin(d))*np.tan(2*np.arcsin(d))-d
    denominator = np.tan(2*np.arcsin(d))
    return numerator/denominator
```

```python
d_array = np.arange(-0.95,1,0.05)
x_array = []
for d in d_array:
    if d != 0:
        x_array.append(cross_value(d))
    else:
        x_array.append(float(0))
```

```python
%matplotlib inline
```

```python
for i in range(len(d_array)):
    x_plot = [0,np.cos(np.sin(d_array[i])),x_array[i]]
    y_plot = [d_array[i],d_array[i],0]
    plt.plot(x_plot,y_plot,'#808080')
    
```
